```python
import pandas as pd
from langdetect import detect, LangDetectException

df = pd.read_csv('../clean.csv')
```

```python
df = df.dropna(subset='text')
```

```python
df
```

```python
def is_english(text):
    try:
        return detect(text) == 'en'
    except LangDetectException:  # Handling cases where language detection may fail
        return False
```

```python
from tqdm import tqdm
tqdm.pandas(desc="Detecting English Rows")
df['is_english'] = df['text'].progress_apply(is_english)
filtered_df = df[df['is_english']]

# Drop the 'is_english' column if not needed
filtered_df = filtered_df.drop(columns=['is_english'])
filtered_df
```

```python
filtered_df = filtered_df.drop(columns=['Unnamed: 0'])
```

```python
filtered_df.to_csv('../clean.csv')
```
